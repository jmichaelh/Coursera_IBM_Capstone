#### Cluster Furnture Store Data 

```python
kclusters = 9

atl_clustering = atl_furniture_group.drop(['Neighborhoods'], 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(atl_clustering)

kmeans.labels_[1:121]
```
